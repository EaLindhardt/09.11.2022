```python
def rens_ord(text_0):
    text_1 = text_0.replace("\n"," ")
    text_2 = text_1.replace("."," ")
    text_3 = text_2.replace(","," ")
    text_4 = text_3.replace(":"," ")
    text_5 = text_4.replace("*"," ")
    text_6 = text_5.replace("–"," ")
    text_7 = text_6.replace("'"," ")
    text_8 = text_7.replace("”"," ")
    text_9 = text_8.replace("-"," ")
    text_10 = text_9.replace(";"," ")
    text_11 = text_10.replace("!"," ")
    text_ren = text_10.replace("?"," ")
    text_lav = text_ren.lower()
    text_token = text_lav.split()
    return text_token

salmer2003["RensetOrd"] = salmer2003.KunTekst.apply(rens_ord)
salmer2003
```

```python
kun_tekst = []
kategorier = []
nummer = []

for fil in os.scandir(r"C:\Users\au576018\Documents\Data\Psalme-bog_for_kirke_og_hjem"):
    with open (fil, encoding = "utf8") as f:
        salme = f.read()
        salme_linjedelt = salme.split('\n')
        kategorier.append(salme_linjedelt[0])
        nummer.append(salme_linjedelt[1])
        indeks = 0
        for i, ltr in enumerate(salme):
            if ltr.isdigit():
                indeks = i
                break
        salme = salme[indeks::]
        salme = salme.replace('\n', ' ')
        kun_tekst.append(salme)
        
kun_tekst_rå_tekst = [''.join([i for i in salmetekst if not i.isdigit()]) for salmetekst in kun_tekst]

salmer1892 = pd.DataFrame(nummer, columns = ['Nummer'])
salmer1892['Kategorier'] = kategorier
salmer1892['RåTekst'] = kun_tekst_rå_tekst
salmer1892['RensetOrd'] = salmer1892.RåTekst.apply(rens_ord)
salmer1892
```

```python
def står_ord_i_ordbog(opslag, nyt_ord):
    values = ordbog.get(opslag)
    if values == None:
        ordbog[opslag] = [nyt_ord]
        return opslag, "og", nyt_ord, "er tilføjet til ordbogen"
    
    if nyt_ord in values:
        return "Opslag i ordbogen"
        
    if opslag in ordbog.keys() and nyt_ord not in values:
        ordbog[opslag] += [nyt_ord]
        return nyt_ord, "er tilføjet til ordbogen ved", opslag
    
```

# SpaCy og Dependency parsing

Jeg vil gerne se hvilke salmer der indeholder et specifikt ord

```python
def tekster_med_ord(lst, wrd):
    indeks_tekster = []
    for i in range(0, len(lst)):
        if wrd in lst[i]:
            indeks_tekster.append(i)
    return indeks_tekster
```

Jeg vil gerne have printet ordene før og efter det pågældende ord, så det er nemmere at se syntaksen og rollefordelingen

```python
def omkringliggende(tekst, wrd):
        
    if wrd in tekst:
        index = tekst.index(wrd)
        before = index - min(index, 5)
        after = index + min( len(tekst) - 1 - index, 5) + 1 
        
    print(tekst[before:after], sep=",")
```
